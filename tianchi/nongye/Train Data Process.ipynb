```python
from osgeo import gdal
from PIL import Image
import os
```

```python
img1path = "D:/kaggle/nongye/jingwei_round1_train_20190619/jingwei_round1_train_20190619/image_1.png"
n = os.path.basename(img1path)[:-4]
label1name = "D:/kaggle/nongye/jingwei_round1_train_20190619/jingwei_round1_train_20190619/" + n + "_label.png"
```

```python
dslb = gdal.Open(label1name)
ds = gdal.Open(img1path)
```

```python
wx = ds.RasterXSize # 47161
wy = ds.RasterYSize # 50141
```

```python
stx = 0
sty = 0
step = 400
outsize = 768
nullthresh = outsize * outsize * 0.7
cx = 0
cy = 0
```

```python
while cy + outsize < wy:
    cx = 0
    while cx + outsize < wx:
        img = ds.ReadAsArray(cx, cy, outsize, outsize)
        img2 = img[:3, :, :].transpose(1, 2, 0)
        if (img2[:, :, 0]==0).sum() > nullthresh:
            cx += step
            print('blank...', cx, cy)
            continue
        
        img2 = Image.fromarray(img2, 'RGB')
        img2.save('D:/kaggle/nongye/jingwei_round1_train_20190619/train/' + n + '_{}_{}.bmp'.format(cx, cy))
        
        img = dslb.ReadAsArray(cx, cy, outsize, outsize)
        img = Image.fromarray(img).convert('L')
        img.save('D:/kaggle/nongye/jingwei_round1_train_20190619/label/' + n + '_{}_{}.bmp'.format(cx, cy))
        
        cx += step
    cy += step
```
